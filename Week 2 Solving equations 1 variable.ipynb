```python
import sympy
from sympy import symbols
from sympy.solvers import solve

x = symbols('x')

# Put the equation here
eq = x - 2
print("x = ", solve(eq, x))

eq = x**2 - 4
print("x = ", solve(eq, x))

eq = 2*x - 4
print("x = ", solve(eq, x))
```

```python
from sympy import *

x, y = var('x y')

# first equation set equal to zero, ready to solve
first = 2*x + 10

#Sympy syntax for equation equal to y, ready to factor
eq1 = Eq(first, y)

# Sympy solve for x
sol = solve(eq1, x)

# show factored results
print("x = ", sol[0])
```

# Explaining how each function works

```python
# converts string input (including fractions) to float
def string_frac(in_string):
  if "/" in in_string:
    nd = in_string.split("/")
    n = float(nd[0])
    d = float(nd[1])
    ans = n/d
    return ans
  else:
    ans = float(in_string)
    return ans

#simplest one-step addition
def one_step_add():
  import random
  # display problem
  a = random.randint(-4,10)
  b = random.randint(2,24)
  print("x + ", a, " = ", b)
  ans = float(input("x = "))
  answer = b-a
  if ans== answer:
    print("Correct! \n")
  else:
    print("Try again")
    print("The correct answer is ", answer, "\n")

# ib-step addition with negative numbers
def one_step_subtract():
  import random
  a = random.randint(-19,-1)
  b = random.randint(2,24)
  print(a, " + x = ", b)
  ans = float(input("x = "))
  answer = b - a
  if ans==answer:
    print("Correct! \n")
  else:
    print("Try again")
    print("The correct answer is ", answer, "\n")

# one-step multiply
def one_step_mult():
  # uses string_frac(<input string>)
  import random
  a = random.randint(1,11)
  b = random.randint(2,24)
  print(a, "x = ", b)
  ans_in = (input("x = "))
  answer = b/a
  if string_frac(ans_in) == answer:
    print("Correct! \n")
  else:
    print("Try again \n")
    print("The correct answer is ", answer, "\n")

def main():
  one_step_mult()

if __name__ == "__main__":
  main()

```

# Factoring and solving for a variable

```python
from sympy import *
import math

# Identify all variables
a, b, c, d, x = var('a b c d x')

# Left and rights sides of the equal sign
left = 0
right = a*x**2 + b*x + c

# Variable to solve for
variable = x

# Sympy equation left = right
eq1 = Eq(left, right)

# Sympy solve for that variable
sol = solve(eq1, variable)

# Show factored results
for s in sol:
  print(variable, " = ", s)
```
