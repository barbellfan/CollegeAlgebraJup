# Fractions and decimals

Converting a repeating decimal back to a fraction:

Starting with 0.3333. Subtract it from 10x, and reduce:

$ 10x = 3.333\overline{3} $

$ - x = 0.333\overline{3} $
<hr>
$ 9x = 3 $

$ x = \frac{3}{9}  = \frac{1}{3} $



What if you have someting like 9.090909? Since there are two digits that repeat, subtract it from *100*x, and reduce:

$ 100x = 9.0909\overline{09} $

$ - x = 0.0909\overline{09} $
<hr>
$ 99x = 9 $

$ x = \frac{9}{99} = \frac{1}{11} $

# Positive and negative exponents

```python
print(10**1)
print(10**2)
print(10**3)

print(10**0)

print(10**-1)
print(10**-2)
print(10**-3)
```

# Cast input as a `float` or `int` variable

```python
text = input("Enter a number: ")
num = float(text)
print(num + 4)
```

```python
text = input("Enter a number: ")
num = int(text)
print(num + 4)
```

# Now put it all together

```python
# Get string input, which will include a decimal point (less than 1)
digits = input("Enter a decimal number to convert: ")

# Get the number of decimal places as an integer
exponent = int(len(digits)) - 1

# Convert the input to a float number
n = float(digits)

# Use the exponent to get the numerator
numerator = int(n * 10**exponent)

# Use the exponent to the the denominator
denominator = 10**exponent

# Percent is the first two decimal places
percent = n * 100

# Output
print("The decimal is ", n)
print("The fraction is ", numerator, "/", denominator)
print("The percent is ", percent, " %")
```
