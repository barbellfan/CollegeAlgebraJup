# Graphing Systems

Graph two lines on one graph. Where these two lines cross is the solution.

For example, you have these two equations:

$y = 3x + 10$

$y = 4x$

Estimate. Say you guess that x = 10. Plug x into these equations:

$40 = 3(10) + 10$

$40 = 4(10)$

So 10 works.

Example: use numpy library insteaed of a loop.

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10

# define how many points to plot
points = 10 * (xmax - xmin)

# define the array of x values once
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax]) # window size
plt.plot([xmin, xmax],[0, 0], 'b') # blue x axis
plt.plot([0, 0], [ymin, ymax], 'b') # blue y axis

# line 1
y1 = 2*x
plt.plot(x, y1)

# line 2
y2 = x**2 - 3
plt.plot(x, y2)

plt.show()
```

# Extra Problems

Shading and graphing inequalities

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10

# define how many points to plot
points = 2 * (xmax - xmin)

# define the array of x values once
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax]) # window size
plt.plot([xmin, xmax],[0, 0], 'b') # blue x axis
plt.plot([0, 0], [ymin, ymax], 'b') # blue y axis

# line 1
y1 = x + 6
plt.plot(x, y1)
plt.fill_between(x, y1, ymax, facecolor='red')

# line 2
y2 = x + 3
plt.plot(x, y2)
plt.fill_between(x, y2, y1, facecolor='yellow')

# line 3
y3 = x - 1
plt.plot(x, y3)
plt.fill_between(x, y3, y2, facecolor='green')

# line 4
y4 = x -4
plt.plot(x, y4)
plt.fill_between(x, y4, y3, facecolor='blue')

plt.show()
```

Interactive graphs. The ipywidgets library only works with jupyterhub. I'm using jupyter-lab, so it won't work. The jupyterhub software is a whole service that runs in, like k8ts, so I'm not going to install it.

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

# Define the graphing function
def f(m, b, zoom):
    xmin = -zoom
    xmax = zoom
    ymin = -zoom
    ymax = zoom
    points = 10 * xmax - xmin
    x = np.linspace(xmin, xmax, points)

    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0]) # black x axis
    plt.plot([0,0], [ymin, ymax]) # black y axis
    plt.title('y = mx + b')

    plt.plot(x, m*x+b)
    plt.show()
    
# set up the sliders
interactive_plot = interactive(f, m=(-9, 9), b=(-9, 9), zoom=(1, 100))
interactive_plot
```
